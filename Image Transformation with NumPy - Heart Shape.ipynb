Create a variable heart_image to representing a 7x7 matrix of RGB values and a function .show_image() to reveal the photo heart_image represents.

Inside .show_image() have two parameters:
<br>1) image: a NumPy array of RGB values
<br>2) name_identifier: a string to title your photo

```python
import numpy as np
import matplotlib.pyplot as plt

heart_img = np.array([[255,0,0,255,0,0,255],
                      [0,255/2,255/2,0,255/2,255/2,0],
                      [0,255/2,255/2,255/2,255/2,255/2,0],
                      [0,255/2,255/2,255/2,255/2,255/2,0],
                      [255,0,255/2,255/2,255/2,0,255],
                      [255,255,0,255/2,0,255,255],
                      [255,255,255,0,255,255,255]])

def show_image(image, name_identifier):
    plt.imshow(image, cmap="gray")
    plt.title(name_identifier)
    plt.show()
```

Show heart image.

```python
show_image(heart_img, "Heart Image")
```

Show inverted heart image.
<br> We can substracts each value in heart_img from 255.

```python
inverted_heart_img = 255 - heart_img
show_image(inverted_heart_img, "Inverted Heart Image")
```

Let’s do another image transformation. This time let’s plot a rotation of heart_img. Create a new variable called rotated_heart_img that swaps the rows and columns of heart_img.

```python
rotated_heart_img = heart_img.T
show_image(rotated_heart_img, "Rotated Heart Image")
```

Solve for the matrix that creates heart_img from an random image.

```python
random_img = np.random.randint(0,255, (7,7))
show_image(random_img, "Random Image")
```

Use NumPy’s linalg.solve() method.

```python
x = np.linalg.solve(random_img, heart_img)
show_image(x, "X")
```

Matrix multiplication using NumPy.

```python
solved_heart_img = random_img@x
show_image(solved_heart_img,"Solved Heart Image")
```
